```python
import pandas as pd

layout = pd.read_clipboard()
```

```python
layout
```

```python
plate = aq.Collection.new(
    object_type=aq.ObjectType.find_by_name('96 Well PCR Plate')
)

for r, (_, row) in enumerate(layout.iterrows()):
    
    for c, sid in enumerate(row.values):
        c = c-1
        if isinstance(sid, str) and 'UWBF' in sid:
            sid = sid.split('UWBF_')[-1]
            plate.sample_id_matrix[r, c] = int(sid)
```

```python
matrix = []

for row in plate.sample_id_matrix.data:
    row_ids = []
    for x in row:
        if x is not None:
            row_ids.append(x.part.sample_id)
        else:
            row_ids.append(None)
    matrix.append(row_ids)

import numpy as np

pd.DataFrame(matrix)
```
