```python
import seaborn as sns
from matplotlib import pylab as plt
import pandas as pd

df = pd.DataFrame({
    'Researcher Hours per Design': [34.28571429, 40, 2.6, 0.875],
    'Cycles': [16, 6, 5, 1],
    'Program Phase': ['Pre-SD2\n(2015-2017)', 'Phase 1\n(2018-2019)', 'Phase 2\n(2019-2020)', 'Phase 3\n(2021)']
})

fig, ax = plt.subplots()

sns.lineplot(data=df, x='Program Phase', y='Researcher Hours per Design', markers='o', color='k', linewidth=3, ax=ax)
ax.set_xlabel('Program Phase', fontsize=12)
ax.set_ylabel('Hours Planning per Design', fontsize=12)

ax2 = ax.twinx()

color = 'firebrick'
sns.lineplot(data=df, x='Program Phase', y='Cycles', markers='o', linewidth=3, ax=ax2, color=color)
ax2.set_ylabel('No. re-work cycles per Design', fontsize=12, color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel('Build Errors', fontsize=12)

plt.tight_layout()
fig.savefig("BuildTimeReduction.png", format='png', dpi=400)
```

```python
import seaborn as sns
from matplotlib import pylab as plt
import pandas as pd

df = pd.DataFrame({
    'Designs Completed Per Month': [1.94, 0.5, 12.5, 8],
    'Build Complexity': [30, 8, 36, 70],
    'Program Phase': ['Pre-SD2\n(2015-2017)', 'Phase 1\n(2018-2019)', 'Phase 2\n(2019-2020)', 'Phase 3\n(2021)']
})

fig, ax = plt.subplots()

sns.lineplot(data=df, x='Program Phase', y='Designs Completed Per Month', color='b', markers='o', linewidth=3, ax=ax)
ax.set_xlabel('Program Phase', fontsize=12)
ax.set_ylabel('Designs Per Month', fontsize=12)
# plt.show()
ax2 = ax.twinx()

color = 'green'
sns.lineplot(data=df, x='Program Phase', y='Build Complexity', markers='o', linewidth=3, ax=ax2, color=color)
ax2.set_ylabel('Build Complexity', fontsize=12, color=color)
ax2.tick_params(axis='y', labelcolor=color)


plt.annotate('automation\ndifficulties' ,xy=(1,10),xytext=(0.2,30),
             arrowprops=dict(arrowstyle='->',lw=1.5))
ax2.set_ylabel('Build Complexity', fontsize=12)
plt.tight_layout()
fig.savefig("BuildSpeedComplexity.png", format='png', dpi=400)
```
